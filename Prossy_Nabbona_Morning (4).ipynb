```python
# Bargraph
pd.crosstab(df.age,df.target).plot(kind='bar',figsize=(20,6))
plt.title('Heart disease frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# data visulasation for heart disease frequency for sex
pd.crosstab(df.sex,df.target).plot(kind='bar',figsize=(20,6), color=['#FFE5C365', '#FFC9DEF5'])
plt.title('Heart disease frequency for Ages')
plt.xlabel('Sex (0=Female, 1=Male)')
plt.legend(['Have disease', 'Doesnot have diesease'])
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.show()
```
